```python
import pandas as pd #pandas library
#read iris dataset from UCI database
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data", 
                 names = [ "sepal_length", "sepal_width", "petal_length", "petal_width", "class" ])
print(df.info()) #information about iris columns
print(df.describe()) #elaborate details and values
```

```python
#define marker shapes by class
import matplotlib.pyplot as plt
marker_shape = ['.', '^', '*']
#then plot the scatterplot
ax = plt.axes()
for i, species in enumerate(df['class'].unique()):
  species_data = df[df['class'] == species]
  species_data.plot.scatter(x='sepal_length', y='sepal_width', marker=marker_shape[i], s=100, title="Sepal width vs length by species", label=species, figsize=(10, 7), ax=ax)
```

```python
df['petal_length'].plot.hist(title="Histogram of petal length")
```

```python
df.plot.box(title="Boxplot of sepal length & width and petal length & width")
```

```python
import numpy as np
import pandas as pd
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data", 
                 names = [ "sepal_length", "sepal_width", "petal_length", "petal_width", "class" ])
random_index = np.random.choice(df.index, replace=False, size=10)
# set the sepal length values of these rows to be none
df.loc[random_index, 'sepal_length']=None
```

```python
print(df.isnull().any())
```

```python
print("Number of rows before deleting: %d" % (df.shape[0]))
df2 = df.dropna()
print("Number of rows after deleting: %d" % (df2.shape[0]))
```

```python
df.sepal_length = df.sepal_length.fillna(df.sepal_length.mean()) #repopulate database to fill up deleted values
```
